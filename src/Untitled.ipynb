```python
def floodfunc(x,y,visited,iterations):
    if iterations>=40:
        return
    visited[x][y] = 1
    img[x][y] = 1
    if x+1<= img.shape[0] and img[x+1][y]!=0 and visited[x+1][y]!=0:
        floodfunc(x+1,y,visited, img, iterations)
    if y+1<= img.shape[1] and img[x][y+1]!=0 and visited[x][y+1]!=0:
        floodfunc(x,y+1, visited, img, iterations)
    if x-1>=0 and img[x-1][y]!=0 and visited[x-1][y]!=0:
        floodfunc(x-1,y, visited, img, iterations)
    if y-1>=0 and img[x][y-1]!=0 and visited[x][y-1]!=0:
        floodfunc(x,y-1, visited, img, iterations)
```

```python
def func(x,y,visited, m,img):
    visited[x][y] =1
    m.append((x,y))
    if x+1<= img.shape[0] and img[x+1][y]==0 and visited[x+1][y]!=0:
        visited,m = func(x+1,y,visited,m, img)
    if y+1<= img.shape[1] and img[x][y+1]==0 and visited[x][y+1]!=0:
        visited,m = func(x,y+1, visited,m, img)
    if x-1>=0 and img[x-1][y]==0 and visited[x-1][y]!=0:
        visited,m = func(x-1,y, visited,m, img)
    if y-1>=0 and img[x][y-1]==0 and visited[x][y-1]!=0:
        visited,m = func(x,y-1, visited,m, img)
    return visited,m
```
